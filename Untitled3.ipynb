```python
import nltk
from nltk.tokenize import word_tokenize
url="https://www.gutenberg.org/cache/epub/76500/pg76500.txt"
response=request.urlopen(url)
raw=response.read().decode('utf-8')
tokens=word_tokenize(raw)
```
