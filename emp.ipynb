```python
class emp:
    emps=0
    raise_amt=1.04
    def __init__(self,fn,ln,pay):
        self.fn=fn
        self.ln=ln
        self.pay=pay
    emps+-1
    def apply_raise(self):
        self.pay=int(self.pay*emp.raise_amt)
```

```python
class mypar:
    def greet(self):
        print("Hello,parent class")

class mych(mypar):
    def greet(self):
        super().greet()
        print("Hello,child class")
obj=mych()
obj.greet()
```
